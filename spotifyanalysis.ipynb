```python
daily_counts = df_world['release_date'].value_counts().sort_index()
daily_positions = df_world.groupby('release_date')['position'].mean().sort_index()

# Grafiği oluşturalım
plt.figure(figsize=(10, 6))
plt.plot(daily_counts.index, daily_counts.values, marker='o', label='Number of Songs')
plt.plot(daily_positions.index, daily_positions.values, marker='o', label='Avarage Ranking')
plt.xlabel('Release Date')
plt.ylabel('Number of Songs / Avarage Ranking')
plt.legend()
plt.title('Number of songs with respect to its release date - Avarage Ranking')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
